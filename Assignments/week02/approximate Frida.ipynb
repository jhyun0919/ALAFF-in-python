```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
A = cv2.imread("Frida.jpg", 0)

# summarize some details about the image
print(type(A))
print(A.shape)

# show the image
plt.imshow(A, cmap='gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
k = 20

# svd
U, s, Vh = np.linalg.svd(A)

# reconstruct the original matrix
m, n = U.shape[0], Vh.shape[0]

Sigma = np.zeros((m, n))
for i in range(k):
    Sigma[i, i] = s[i]

B = np.dot(U, np.dot(Sigma, Vh))
```

```python
# summarize some details about the converted data
print(type(B))
print(B.shape)

# show image
plt.imshow(B, cmap='gray')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
